# Column Name & Description

- **VIN (1-10)**: Vehicle Identification Number (partial, first 10 chars). Unique ID for each vehicle. Use for identifying duplicates or tracking individual entries.

- **County**: The county where the vehicle is registered. Useful for regional analysis.

- **City**: The city where the vehicle is registered. Useful to find top EV cities.

- **State**: State abbreviation (always "WA" = Washington).

- **Postal Code**: Area zip code. Can be used for location-level mapping or grouping.

- **Model Year**: The year the EV was manufactured. Helpful in finding trends over years.

- **Make**: The brand/company name (e.g., TESLA, FORD, NISSAN).

- **Model**: Specific model name (e.g., Model 3, LEAF, Mustang Mach-E).

- **Electric Vehicle Type**: EV type:
  - Battery Electric Vehicle (BEV) = fully electric
  - Plug-in Hybrid Electric Vehicle (PHEV) = electric + fuel

- **Clean Alternative Fuel Vehicle (CAFV) Eligibility**: Whether the vehicle is eligible for the clean fuel program. Useful for policy eligibility analysis.

- **Electric Range**: Maximum distance (in miles/km) the vehicle can travel on a full electric charge. Used to compare efficiency.

- **Base MSRP**: Manufacturer Suggested Retail Price (initial cost of vehicle). Great for price comparison and budgeting.

- **Legislative District**: WA state legislative district code. Useful for political/geographical mapping.

- **DOL Vehicle ID**: Department of Licensing internal vehicle ID. Just a unique identifier (not useful for EDA).

- **Vehicle Location**: Coordinates (longitude, latitude) where vehicle is registered. Useful for mapping on map/chart.

- **Electric Utility**: The company supplying electricity to that region (e.g., PACIFICORP, PUGET SOUND ENERGY).

- **2020 Census Tract**: Geographical block code from 2020 US Census. Used in mapping/demographics.


# Import library and Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
%matplotlib inline
warnings.filterwarnings('ignore')

df = pd.read_csv("electric_data.csv")
```

 ## This horizontal bar chart visualizes the top 10 electric vehicle (EV) brands registered in the state of Washington.
### Each bar represents the number of vehicles registered under a specific brand.

 - TESLA leads by a significant margin, reflecting its strong presence and popularity in the EV market.

- Chevrolet and Nissan follow as the next most popular EV brands, showing consistent adoption.

- Other notable brands include Ford, BMW, and Toyota, indicating a diverse brand mix in the EV ecosystem.

##### The chart highlights the dominance of certain manufacturers and provides insight into consumer preferences in Washington's growing electric vehicle market.



```python
top_makes = df["Make"].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_makes.values, y=top_makes.index, palette="viridis")
plt.title("Top 10 EV Brands in Washington")
plt.xlabel("Number of Vehicles")
plt.ylabel("Brand")
plt.show()
```

# This line chart illustrates the year-wise trend of electric vehicle (EV) registrations in Washington state.
### The X-axis represents the model year of the EVs, while the Y-axis shows the number of vehicles registered.

- **Key Observations:**

* EV registrations remained low until around 2015â€“2017

* A sharp increase is seen after 2018, with the highest spike between 2021â€“2023

* The upward trend reflects growing EV adoption, likely due to technological improvements, policy incentives, and increased awareness

```python
#EVs Registered Per Year 

year_count=df["Model Year"].value_counts().sort_values()
plt.figure(figsize=(10, 6))
sns.lineplot(x=year_count.index, y=year_count.values, marker="o", color="blue")
plt.title("Electric Vehicle Registrations by Year")
plt.xlabel("Model Year")
plt.ylabel("Number of EVs")
plt.grid(True)
plt.show()
```

# This bar chart presents the top 10 electric vehicle brands in terms of their average electric driving range.

- The X-axis shows the average electric range (in kilometers) for each brand.

- The Y-axis lists the brand names.

```python
#Average Electric Range by Top 10 Brands

avg_range_by_make = df.groupby("Make")["Electric Range"].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_range_by_make.values, y=avg_range_by_make.index, palette="magma")
plt.title("Top 10 Brands by Average Electric Range")
plt.xlabel("Average Range (km)")
plt.ylabel("Brand")
plt.show()
```

# This bar chart displays the distribution of electric vehicles based on their CAFV (Clean Alternative Fuel Vehicle) Eligibility status.

- The X-axis shows the CAFV eligibility categories

- The Y-axis shows the number of vehicles in each category

```python
# Vehicle Type Distribution

cafv = df["Clean Alternative Fuel Vehicle (CAFV) Eligibility"].value_counts()
plt.figure(figsize=(16, 5))
sns.barplot(x=cafv.index, y=cafv.values, palette="viridis")
plt.title("EV Type Distribution")
plt.ylabel("Number of EVs")
plt.xlabel("EV Type")
plt.show()
```

# This horizontal bar chart represents the top 10 counties in Washington State with the highest number of registered electric vehicles (EVs).

- The X-axis shows the number of EVs registered

- The Y-axis shows the county names

```python
#Top 10 Counties by EV Registrations

top_county=df["County"].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_county.values, y=top_county.index, palette="coolwarm")
plt.title("Top Counties by EV Registrations")
plt.xlabel("Number of Vehicles")
plt.ylabel("County")
plt.show()
```
